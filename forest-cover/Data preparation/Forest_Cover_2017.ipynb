```python
# Authored by Safa Kaysir with support from Karen Scott

def prepare_forest_cover():
  from google.colab import files 
  import pandas as pd 
  # forestcover = pd.read_csv('https://drive.google.com/uc?id=1oAVeevooBzbLwMdczMM1eIxlDeH4_8wo')
  forestcover = pd.read_csv('https://drive.google.com/uc?id=1KczTw04OwIbKYKz-Jx1MeJN87THq0_R_')

  #renaming the unnamed column 
  forestcover.rename(columns ={'Unnamed: 1':'Region'},inplace =True)

  # remove all records after row 683 - these are individual countries
  forestcover = forestcover.iloc[:682,:]

  # remove the total records
  forestcover = forestcover[(~forestcover['Region'].str.contains('Total'))]

  # isolate records that show Forest cover (thousand hectares) in series column and are for 2017
  forestcover = forestcover[(forestcover['Series'] == 'Forest cover (thousand hectares)')&(forestcover['Year']== 2017)]

  # remove records that are for the continents rather than regions
  forestcover = forestcover.query("Region not in ['Africa','Americas','Europe','Asia','Oceania']")

  # isolate just the Region and Value columns
  forestdata = forestcover[['Region','Value']]

  display(forestdata)

  # transpose the code and save to json file
  forestdata = forestdata.transpose()    
  forestdata.to_json('Forest-Cover-proportion-2017')
  files.download('Forest-Cover-proportion-2017')

prepare_forest_cover()

```
